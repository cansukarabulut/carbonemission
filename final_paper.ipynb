```python
import pandas as pd
import plotly.express as px
from plotnine import *

# Fuel Consumption dataset
Fuel_Consumption= pd.read_csv('https://raw.githubusercontent.com/cansukarabulut/carbonemission/main/Fuel_Consumption.csv')
```

```python
df=pd.read_csv("Fuel_Consumption.csv")
df
```

```python
df.groupby("MAKE")["EMISSIONS"].mean()
```

```python
import matplotlib.pyplot as plt

result = df.groupby("MAKE")["EMISSIONS"].mean()
result.plot(kind='bar', width=0.5, align='edge')

plt.xlabel('Make')
plt.ylabel('Mean Emissions')
plt.title('Mean Emissions by Make')
plt.xticks(ticks=range(len(result.index)), labels=result.index, rotation=45, ha='right')

plt.tight_layout()
every_nth = 0.5
for n, label in enumerate(plt.gca().xaxis.get_ticklabels()):
    if n % every_nth != 0:
        label.set_visible(False)
for label in plt.gca().xaxis.get_ticklabels():
    label.set_visible(True)

plt.show()
```

```python
result = df.groupby("MAKE", as_index = False).agg(average_emission = ('EMISSIONS', 'mean'))
result.sort_values('average_emission', ascending = False, inplace = True)
```

#Buggatti has significant role

```python
result
```

```python

p= px.bar(result, x='MAKE', y='average_emission', text_auto= '.2s')
p.update_layout( title =  'Average Emission by Manufacturer', xaxis_title= 'Manufacturer Company', yaxis_title='Average Emission', plot_bgcolor = 'White' )



```

```python
df
px.bar(df,x='MAKE',y= 'EMISSIONS')
fig = px.bar(df,x='MAKE',y= 'EMISSIONS')
fig.update_layout(plot_bgcolor= 'white',title='EMISSIONS BY MAKE' ,
                  width=950, height=500, xaxis_title = "MAKE", yaxis_title = "emissions")
fig.update_traces(textposition='outside', texttemplate='%{value:.3s}')
fig.update_xaxes(type='category')
fig.show()
```

```python
result = df.groupby("MAKE")["EMISSIONS"].mean().reset_index()

fig = px.bar(result, x='MAKE', y='EMISSIONS')

fig.update_layout(plot_bgcolor='white', title='Mean Emissions by Make', width=950, height=500, xaxis_title="MAKE", yaxis_title="emissions")
fig.update_traces(textposition='outside', texttemplate='%{value:.3s}')

fig.show()
```
